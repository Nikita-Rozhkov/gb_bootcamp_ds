$\sigma^{\prime}(x)=\frac{0\cdot(1+e^{-x})-1\cdot(1+e^{-x})^{\prime}}{(1+e^{-x})^2}=\frac{-(-e^{-x})}{(1+e^{-x})^2}=\frac{e^{-x}}{(1+e^{-x})^2}=\frac{1}{1+e^{-x}}\cdot\frac{e^{-x}}{1+e^{-x}}=\sigma(x)\cdot\frac{e^{-x}}{1+e^{-x}}=\sigma(x)\cdot\frac{1+e^{-x}-1}{1+e^{-x}}=\sigma(x)\cdot(1-\frac{1}{1+e^{-x}})=\sigma(1-\sigma(x))$
