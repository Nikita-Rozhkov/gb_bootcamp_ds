# Возьмём несколько признаков

```python
features = [
    'store',
    'dayofweek',
    'customers',
    'open',
    'promo',
]

target = ['sales']
```

# Практическое задание

В sklearn есть встроенные («игрушечные») наборы данных. Возьмём для примера один из них — load_boston(), в нём содержится информация о ценах на недвижимость в Бостоне.

Посмотрим его описание

```python
from sklearn.datasets import load_boston

boston_data = load_boston()
print(boston_data['DESCR'])
```

Загрузим данные о признаках в "x", а целевые значения — в "y"

```python
import pandas as pd
```

```python
X = pd.DataFrame(boston_data.data, columns=boston_data.feature_names)
y = boston_data.target
```

Посмотрим основную информацию о нём

```python
# Первые 5 строк
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
X_train.index
```

```python
X_train
```

```python
y_train
```

## Обучить модель

```python
from sklearn.linear_model import LinearRegression
lin_model = LinearRegression()
```

```python
lin_model.fit(X_train, y_train)
```

```python
pred_train = lin_model.predict(X_train)
pred_train
```

```python
pred_test = lin_model.predict(X_test)
```

```python
pred_test
```

```python
train_preds = pd.DataFrame(pred_train, columns=['pred'])
train_preds['target'] = y_train
train_preds
```

```python
test_preds = pd.DataFrame(pred_test, columns=['pred'])
test_preds['target'] = y_test
test_preds
```

## Оценить точность модели на обучающей выборке и тестовой

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
model_metrics = pd.DataFrame(index = ['test', 'train'], columns=['mse', 'r2'])
model_metrics['mse'].loc['train'] = mean_squared_error(y_train, pred_train)
model_metrics['r2'].loc['train'] = r2_score(y_train, pred_train)
model_metrics['mse'].loc['test'] = mean_squared_error(y_test, pred_test)
model_metrics['r2'].loc['test'] = r2_score(y_test, pred_test)
```

```python
model_metrics
```
