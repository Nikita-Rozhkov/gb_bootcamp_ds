### Импорт библиотек

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

### Загрузка и обзор данных

```python
data = sns.load_dataset('iris')
data.head()
```

```python
data.info()
```

```python
data.describe()
```

### Получение статистик

```python
# распределение классов
data['species'].value_counts()
```

```python
# средние значения признаков для каждого класса
data.groupby('species').mean()
```

```python
# максимальные значения признаков для каждого класса
data.groupby('species').max()
```

```python
# минимальные значения признаков для каждого класса
data.groupby('species').min()
```

```python
# медианы признаков для каждого класса
data.groupby('species').median()
```

### Матрица корреляции

```python
corr_matrix=data.corr()
corr_matrix
```

```python
plt.figure(figsize=(8, 8))
sns.set(font_scale=1.2)
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap="YlGnBu")
plt.title('Матрица корреляций')
plt.show()
```

### Диаграммы рассеяния и графики распределения признаков

```python
sns.pairplot(data, hue='species', palette='dark')
plt.show()
```

### Box plot

```python
plt.figure(figsize=(8, 8))
sns.set_style('whitegrid')
sns.boxplot(data['species'], data['petal_length'], palette='Set2')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
sns.set_style('whitegrid')
sns.boxplot(data['species'], data['petal_width'], palette='Set1')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
sns.set_style('whitegrid')
sns.boxplot(data['species'], data['sepal_width'], palette='Set3')
plt.show()
```

```python
plt.figure(figsize=(8, 8))
sns.set_style('whitegrid')
sns.boxplot(data['species'], data['sepal_length'], palette='Paired')
plt.show()
```

### Barplot

```python
plt.figure(figsize=(8, 8))
sns.set_style('whitegrid')
sns.barplot(data['species'], data['petal_length'], palette='husl')
plt.show()
```
